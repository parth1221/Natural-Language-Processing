```python
from urllib import request
```

```python
url = "http://www.gutenberg.org/cache/epub/60942/pg60942.txt"
```

```python
response = request.urlopen(url)
```

```python
# decode unicode
raw = response.read().decode('utf8')
```

```python
raw
```

```python
type(raw)
```

```python
len(raw)
```

```python
raw[:75]
```

```python
# Tokenize
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(raw)
```

```python
tokens
```

```python
type(tokens)
```

```python
len(tokens)
```

```python
text = nltk.Text(tokens)
```

```python
words = [w.lower() for w in text]
words
```

```python
# Vocabulary
vocab = sorted(set(words))
```
